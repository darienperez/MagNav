```python
def get_intensity(image, row):
    # Fill in the three lines below
    red_values = image[row, :, 0]
    green_values = image[row, :, 1]
    blue_values = image[row, :, 2]
    
    return [red_values, green_values, blue_values]
```

```python
def get_info(values):
    for i,value in enumerate(values):
        return [i, value]
```

```python
def make_info_list(values, threshold):
    info_list = []
    for i,value in enumerate(values):
        if value <= threshold:
            info_list.append([i,value])
        else:
            pass
        
    return info_list
```
